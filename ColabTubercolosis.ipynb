**¿Qué es la tubercolosis?**

La tuberculosis (TB) es una enfermedad infecciosa causada por la bacteria Mycobacterium tuberculosis. Afecta principalmente a los pulmones, pero también puede afectar otras partes del cuerpo como los riñones, la columna vertebral y el cerebro. Se propaga de persona a persona a través del aire, cuando una persona con TB pulmonar tose, estornuda o habla.

**Métodos de diagnóstico de la tuberculosis:**

**Prueba cutánea de tuberculina (PPD):** Es una prueba común para detectar si una persona ha estado expuesta a la bacteria de la TB. Se inyecta una pequeña cantidad de tuberculina debajo de la piel y se observa la reacción después de 48 a 72 horas.

**Pruebas de sangre para la TB (IGRA):** Estas pruebas, como QuantiFERON-TB Gold Plus o T-SPOT.TB, miden la respuesta inmunitaria a las bacterias de la TB en la sangre. Son más específicas que la PPD y a menudo se usan en personas que han recibido la vacuna BCG.

**Radiografía de tórax:** Se utiliza para buscar signos de TB activa en los pulmones, como infiltrados o cavidades.

**Análisis de esputo (baciloscopia y cultivo):** Una muestra de esputo (flema) se examina bajo un microscopio para buscar la presencia de bacterias de TB. Además, el esputo se puede cultivar en un laboratorio para permitir que las bacterias crezcan, lo que confirma el diagnóstico y permite realizar pruebas de sensibilidad a los medicamentos.

**Pruebas moleculares rápidas (por ejemplo, GeneXpert):** Estas pruebas detectan el ADN de la bacteria de la TB y pueden identificar la resistencia a ciertos medicamentos (como la rifampicina) en pocas horas. Son muy útiles para un diagnóstico rápido y preciso.

**Biopsia:** En casos donde la TB afecta otros órganos, puede ser necesario tomar una muestra de tejido para su análisis.

Es importante destacar que un diagnóstico temprano y un tratamiento adecuado son cruciales para controlar la propagación de la TB y curar al paciente.

**Explorar los valores nulos/faltantes/NA**

```python
# los valores NA se debe a que el iso2 de namibia es "NA", y pandas lo tomo con
# un valor faltante

who.loc[who.country == "Namibia", "iso2"] = "NA"
```

```python
# Las dddd columnas de casos con valores faltantes van a tener un valor defaut de 0

who.fillna(0, inplace=True)
```

```python
# cuantos valores faltantes hoy por columnas en la tabla de pop?
pop.isna().sum()
```

```python
# como puedo convertir columna en filas utilizando pandas?::Existe la función melt()
#buscar la fncion melt en la documentación oficial
#tengo una tabla de esta estructura:

who2 = who.melt(id_vars=["country", "iso2", "iso3", "year"])
who2.head(10)
```

```python
# crear una columna que contenga el genero (Masculino o Femenino)
#crear una función de python que contenga llas reglas

def asignar_genero(valor):
  if "n" in valor:
    return "Masculino"
  else:
      return "Femenino"
```

```python
who2["genero"] = who2["variable"].apply(asignar_genero)
who2.head(15)
```

```python
# the method of diagnosis (rel = relapse, sn = negative pulmonary smear, sp = positive pulmonary smear, ep = extrapulmonary)
def asignar_metodo(valor):
  if "rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapulmonary"
  else:
    return "to define"
```

```python
who2["metodo"] = who2["variable"].apply(asignar_metodo)
who2.head()
```

```python
# age group (014 = 0-14 yrs of age, 1524 = 15-24, 2534 = 25-34, 3544 = 35-44 years of age, 4554 = 45-54, 5564 = 55-64, 65 = 65 years or older).
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoedad"] = who2["variable"].apply(asignar_grupoedad)
who2.head()
```

```python
# remover la columna variable
who2.drop(columns=["variable"], inplace=True)
```

```python
# remover
who2.rename(columns={"value":"cases"}, inplace=True)
```
